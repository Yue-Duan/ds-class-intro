### Exercise 5. Find the mean and standard deviation of math and verbal SAT score for men-only, women-only, and non gender specific universities.

### Exercise 7. Generate a DataFrame for the ratios of (number of employees of specific gender and race/total number of employees) for all race-gender combinations.

```python
df_employee = pd.read_csv('data/employee.csv')
df_employee.tail(10)
```

```python
master_list=[]
distinct=df_employee["race"].drop_duplicates()
race_list=["White","Black","Asian","Hispanic","NaN","Native American"]
gender_list=["Male","Female"]
num_employee=len(df_employee)
for race in race_list:
    for gender in gender_list:
        target_employee=len(df_employee.loc[(df_employee['race'] ==race) & (df_employee['gender'] ==gender)])
        ratio=target_employee/num_employee
        race_gender=race+" "+gender
        row=[race_gender,ratio]
        master_list.append(row)
column_names=["race-gender","ratio"]
output_df = pd.DataFrame(master_list, columns=column_names)
output_df
```

### Exercise 8. Use `pd.melt()` to unpivot table `pv3` to the format of `pv4` in exampel above.

```python
##getting pv3 and pv4 from notebook
emp = pd.read_csv('data/employee.csv')
pv1 = emp.pivot_table(index='dept', columns='gender', 
                      values='salary', aggfunc='mean').round(-3)
pv2 = emp.pivot_table(index='gender', columns='race', 
                      values='salary', aggfunc='mean').round(-3)
pv3 = pv2.reset_index()
pv4 = pv2.stack().reset_index()
```

```python
pd.melt(pv3, id_vars=['gender'], value_vars=['Asian',"Black","Hispanic","Native American","White"],
       var_name='race', value_name='value')
```
